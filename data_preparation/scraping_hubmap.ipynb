```python
import requests
from hubmap_api_py_client import Client
import json
```

```python
client = Client('https://cells.api.hubmapconsortium.org/api/')
#[m for m in dir(client) if m.startswith('select_')]
datasets_ids = list(client.select_datasets().get_list())
ids=list(map(lambda x : x['uuid'],datasets_ids))
```

```python
print(client)
print(len(ids))
```

```python


def find_ancestors(uuid):
  query_dict={
    "query": {
      "bool": {
        "filter": [
          {
            "match": {
              "uuid": uuid
            }
          }
        ]
      }
    }
  }

  #finding all ancestors
  response = requests.post(
  #    'https://search.api.hubmapconsortium.org/search',
      'https://search.api.hubmapconsortium.org/v3/portal/search',
      json = query_dict)
  #print(response.text)
  ancestors=json.loads(response.text)['hits']['hits'][0]['_source']['ancestor_ids']
  return ancestors
```

```python
#be wary, this code took 96m to execute 
graph={}
for uuid in ids:
    ancestors=find_ancestors(uuid)
    ancestors_new=[a for a in ancestors if a not in graph]
    print(uuid + ' '+str(len(ids)) + ' ancestors:' +str(len(ancestors)) + ' new: ' + str(len(ancestors_new)))
    if len(ancestors_new)>0:
        ids.extend(ancestors_new)
        graph[uuid]=ancestors_new
```
